```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Proyectos (1).csv')
df.head()
```

Distribucion de los proyectos por año

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Ejercicio',data=df,palette ='coolwarm')
plt.title('Distribicion de poryecto en un año')
plt.xlabel('Ejercicio')
plt.ylabel('Cantidad de proyectos')
plt.show()

```

Duracion promedio de los proyectos


```python

df['Fecha de Inicio'] = pd.to_datetime(df['Fecha de Inicio'])
df['Fecha de Finalizacion'] = pd.to_datetime(df['Fecha de Finalizacion'])

df['Duracion (Dias)'] = (df['Fecha de Finalizacion'] - df['Fecha de Inicio']).dt.days

df.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Rubro', y='Duracion (Dias)', data=df, estimator=lambda x: x.mean(), palette='magma')
plt.title('Duracion promedio de proyectos')
plt.xlabel('Rubro')
plt.ylabel('Duracion en dias')
plt.xticks(rotation=45)

plt.show()
```

Mapa de calor por tipo de Recurso y rubro

```python
pivot_table=df.pivot_table(
    index ='Tipo de Recurso',
    columns='Rubro',
    aggfunc= 'size',
    fill_value=0

)

plt.figure(figsize=(10,6))
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, annot=True, cmap='Blues')


plt.title('Proyectos por tipo de recurso y rubro')
plt.xlabel('Rubro')
plt.ylabel('Recurs0')


plt.show()
```

Proyectos Comunes por institucion

```python
plt.figure(figsize=(10,6))
df['Convocatoria'].value_counts()[:10].plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='coolwarm')

plt.title('Proyectos por convocatoria')
plt.ylabel('')
plt.show()
```
