```python
import re
import pandas as pd

# Cargar el dataset
df = pd.read_csv('/content/Ejemplo_01_03_Amazon_Sales_Dataset (1).csv')
df.head()
```

```python
def find_phone_number_with_101(text):
    pattern = r'\b101-\d{3}-\d{4}' # 101-903-9622
    result = re.findall(pattern, text) # Buscame el patron/coincidencias (pattern) en el objeto text
    return bool(result)

df_with_101_phone_numbers = df[df['Phone Number'].apply(find_phone_number_with_101)]
df_with_101_phone_numbers.head()
```

```python
#Busqueda de rese√±a con la palabra "Excellent"
def find_reviews_with_excellent(text):
    pattern = r'\bexcellence\b'
    result = re.findall(pattern, text, )  # Use re.IGNORECASE for case-insensitive matching
    return bool(result)


# Apply the function to filter the DataFrame
df_with_excellent_review = df[df['Review'].apply(find_reviews_with_excellent)]

# Display the filtered DataF
df_with_excellent_review.head()
```
