```python
import plotly.express as px
import pandas as pd

df =pd.read_csv('/content/temblores_mexico_50_anios.csv')
df.head()
```

```python
fig = px.scatter_geo(
    df,
    lat='Latitud',
    lon='Longitud',
    hover_name='Ubicación',
    size='Magnitud',
    color='Magnitud',
    animation_frame='Fecha',
    title='Temblores en MX',
    scope='north america'
)

fig.show()


```
